```python
from keras.datasets import cifar10

(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

```python
#save the model after training
import os
model_name = 'cifar10_trained_model_data_augmentation.h5'
```

# **Process data before training**

```python
batch_size = 32
num_classes = 10
epochs = 200
```

```python
import keras
from keras import layers
from keras import models
train_labels = keras.utils.to_categorical(train_labels, num_classes)
test_labels = keras.utils.to_categorical(test_labels, num_classes)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Split data into training and validation
imgaes_train, images_vali, labels_train, labels_vali = train_test_split(train_images, train_labels, test_size=0.2, random_state = 42) 
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

train_datagen.fit(imgaes_train)
train_generator = train_datagen.flow(imgaes_train, 
                                     labels_train,
                                     batch_size=batch_size)

vali_datagen = ImageDataGenerator(rescale=1./255)
vali_generator = vali_datagen.flow(images_vali, 
                                   labels_vali,
                                   batch_size=batch_size)
```

```python
test_images = test_images.astype('float32')
test_images /= 255
```

#**Create convnet**



```python
from keras.layers import Conv2D, MaxPooling2D
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
model = Sequential()

#maybe seperate activation //model.add(Activation('relu')
model.add(Conv2D(32, (3, 3), padding="same", activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D(2, 2))

model.add(Conv2D(32, (3, 3), padding="same", activation='relu'))
model.add(layers.MaxPooling2D(2, 2))

model.add(Conv2D(64, (3, 3), padding="same", activation='relu'))
model.add(layers.MaxPooling2D(2, 2))

model.add(Conv2D(64, (3, 3), padding="same", activation='relu'))
model.add(layers.MaxPooling2D(2, 2))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
#steps_per_epoch = num of training sample / batch size
#validation_steps = TotalvalidationSamples / ValidationBatchSize
model.fit_generator(train_generator,
                    steps_per_epoch=1250,
                    epochs=epochs,
                    validation_data=(images_vali, labels_vali),
                    validation_steps=312)
```

```python
model.save(model_name)
```
